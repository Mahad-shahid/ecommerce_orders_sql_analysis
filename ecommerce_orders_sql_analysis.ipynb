```python
import sqlite3
import pandas as pd
```

```python
# Connect to the SQLite database file (make sure it's in the same folder as the notebook)
conn = sqlite3.connect("retail_store.db")
```

```python
# Check which tables are in the database
tables_query = "SELECT name FROM sqlite_master WHERE type='table';"
pd.read_sql_query(tables_query, conn)
```

```python
pd.read_sql_query("SELECT * FROM Customers LIMIT 5", conn)
```

```python
pd.read_sql_query("SELECT * FROM Products LIMIT 5", conn)
```

```python
pd.read_sql_query("SELECT * FROM Orders LIMIT 5", conn)
```

```python
pd.read_sql_query("SELECT * FROM OrderDetails LIMIT 5", conn)
```

```python
query = """
SELECT 
    COUNT(DISTINCT Orders.OrderID) AS total_orders,
    ROUND(SUM(OrderDetails.TotalPrice), 2) AS total_revenue
FROM Orders
JOIN OrderDetails ON Orders.OrderID = OrderDetails.OrderID;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    Products.ProductName,
    SUM(OrderDetails.Quantity) AS total_sold
FROM OrderDetails
JOIN Products ON OrderDetails.ProductID = Products.ProductID
GROUP BY Products.ProductName
ORDER BY total_sold DESC
LIMIT 5;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    strftime('%Y-%m', Orders.OrderDate) AS Month,
    ROUND(SUM(OrderDetails.TotalPrice), 2) AS MonthlyRevenue
FROM Orders
JOIN OrderDetails ON Orders.OrderID = OrderDetails.OrderID
GROUP BY Month
ORDER BY Month;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    Products.ProductName,
    ROUND(SUM(OrderDetails.TotalPrice), 2) AS TotalRevenue
FROM OrderDetails
JOIN Products ON OrderDetails.ProductID = Products.ProductID
GROUP BY Products.ProductName
ORDER BY TotalRevenue DESC
LIMIT 5;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    Customers.FullName AS CustomerName,
    COUNT(DISTINCT Orders.OrderID) AS OrdersPlaced
FROM Customers
JOIN Orders ON Customers.CustomerID = Orders.CustomerID
GROUP BY Customers.FullName
ORDER BY OrdersPlaced DESC
LIMIT 5;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    ProductName
FROM Products
WHERE ProductID NOT IN (SELECT DISTINCT ProductID FROM OrderDetails);
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    Category,
    ROUND(SUM(OrderDetails.TotalPrice), 2) AS TotalRevenue
FROM Products
JOIN OrderDetails ON Products.ProductID = OrderDetails.ProductID
GROUP BY Category
ORDER BY TotalRevenue DESC;
"""
pd.read_sql_query(query, conn)
```
